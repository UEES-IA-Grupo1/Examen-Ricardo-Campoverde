```python
# Importamos las bibliotecas
import nltk
import spacy
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
```

```python
# Descargamos los recursos necesarios de NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Cargamos el modelo de spaCy
nlp = spacy.load('es_core_news_sm')
```

```python
def analizar_texto(texto):
    """
    Realiza un análisis completo del texto
    """
    print("\n=== ANÁLISIS DE TEXTO ===")

    # 1. Tokenización de Oraciones
    print("\n1. TOKENIZACIÓN DE ORACIONES:")
    oraciones = sent_tokenize(texto)
    for i, oracion in enumerate(oraciones, 1):
        print(f"Oración {i}: {oracion.strip()}")

    # 2. Tokenización de Palabras
    print("\n2. TOKENIZACIÓN DE PALABRAS:")
    palabras = word_tokenize(texto)
    print("Lista de tokens:", palabras)

    # 3. Análisis con spaCy
    print("\n3. ANÁLISIS MORFOLÓGICO CON SPACY:")
    doc = nlp(texto)
    for token in doc:
        print(f"Token: {token.text:15} | Tipo: {token.pos_:10} | Dependencia: {token.dep_:10}")

    # 4. Análisis de Frecuencia
    print("\n4. ANÁLISIS DE FRECUENCIA:")
    palabras = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]
    freq_dist = Counter(palabras)
    print("\nPalabras más frecuentes:")
    for palabra, freq in freq_dist.most_common(5):
        print(f"{palabra}: {freq} veces")

    # 5. Entidades Nombradas
    print("\n5. RECONOCIMIENTO DE ENTIDADES NOMBRADAS:")
    for ent in doc.ents:
        print(f"Texto: {ent.text:20} | Tipo: {ent.label_:10}")

    # 6. Análisis de Estructura
    print("\n6. ESTRUCTURA SINTÁCTICA:")
    for token in doc:
        print(f"Token: {token.text:15} <- Padre: {token.head.text:15} | Relación: {token.dep_}")

# Textos de ejemplo
textos_ejemplo = [
    """
    La inteligencia artificial está revolucionando la tecnología moderna.
    Los sistemas de IA procesan grandes cantidades de datos.
    """,

    """
    El procesamiento de lenguaje natural es una rama importante de la IA.
    Microsoft y Google desarrollan herramientas avanzadas de NLP.
    """
]

# Analizamos cada texto
for i, texto in enumerate(textos_ejemplo, 1):
    print(f"\n{'='*50}")
    print(f"ANÁLISIS DEL TEXTO {i}")
    print(f"{'='*50}")
    analizar_texto(texto)

# Función adicional para análisis específico de entidades
def analizar_entidades(texto):
    """
    Realiza un análisis específico de las entidades en el texto
    """
    print("\n=== ANÁLISIS DETALLADO DE ENTIDADES ===")
    doc = nlp(texto)

    # Clasificar entidades por tipo
    entidades_por_tipo = {}
    for ent in doc.ents:
        if ent.label_ not in entidades_por_tipo:
            entidades_por_tipo[ent.label_] = []
        entidades_por_tipo[ent.label_].append(ent.text)

    # Mostrar resultados
    for tipo, entidades in entidades_por_tipo.items():
        print(f"\nTipo: {tipo}")
        print("Entidades encontradas:", ", ".join(entidades))

print("\nAnálisis adicional de entidades:")
for texto in textos_ejemplo:
    analizar_entidades(texto)
```

```python
def analisis_sentimientos_simple(texto):
    """
    Realiza un análisis de sentimientos básico usando diccionarios de palabras
    """
    # Diccionarios de palabras de sentimientos en español
    palabras_positivas = {
        'bueno', 'excelente', 'feliz', 'alegre', 'maravilloso', 'genial', 'encantado',
        'positivo', 'fantástico', 'increíble', 'satisfecho', 'contento', 'agradable',
        'perfecto', 'espectacular', 'favorable', 'optimista', 'brillante', 'estupendo',
        'extraordinario', 'amoroso', 'divertido', 'eficiente', 'éxito','limpio', 'cómodo', 'acogedor', 'tranquilo', 'céntrico',
        'espacioso', 'moderno', 'amable', 'atento', 'relajante'

    }

    palabras_negativas = {
        'malo', 'terrible', 'triste', 'enojado', 'pésimo', 'horrible', 'deficiente',
        'negativo', 'desagradable', 'fatal', 'fracaso', 'peor', 'mediocre', 'molesto',
        'desastroso', 'irritante', 'frustrado', 'decepcionante', 'inferior', 'peligroso',
        'preocupante', 'problemático', 'insatisfecho', 'descontento', 'sucio', 'ruidoso', 'pequeño', 'viejo',
        'descuidado', 'incómodo', 'lento', 'frío'

    }

    try:
        # Convertimos a minúsculas y tokenizamos
        palabras = word_tokenize(texto.lower())

        # Contamos palabras positivas y negativas
        contador_positivas = sum(1 for palabra in palabras if palabra in palabras_positivas)
        contador_negativas = sum(1 for palabra in palabras if palabra in palabras_negativas)

        # Determinamos el sentimiento
        # Determinamos el sentimiento
        diferencia = contador_positivas - contador_negativas

        if abs(diferencia) <= 1:
            sentimiento = 'neutral'
            emoticon = '➖'
        elif diferencia > 1:
            sentimiento = 'positivo'
            emoticon = '⭐'
        else:
            sentimiento = 'negativo'
            emoticon = '❌'


        # Calculamos la intensidad (opcional)
        total_palabras = len(palabras)
        intensidad_sentimiento = (contador_positivas + contador_negativas) / total_palabras if total_palabras > 0 else 0

        # Mostramos resultados
        print(f"\nReseña de Hotel Analizada: '{texto}'")
        print(f"Calificación General: {sentimiento} {emoticon}")
        print(f"Palabras positivas encontradas: {contador_positivas}")
        print(f"Palabras negativas encontradas: {contador_negativas}")
        print(f"Nivel de satisfacción: {intensidad_sentimiento:.2%}")


        return {
            'sentimiento': sentimiento,
            'positivas': contador_positivas,
            'negativas': contador_negativas,
            'intensidad': intensidad_sentimiento
        }

    except Exception as e:
        print(f"Error en el análisis: {e}")
        return None

# Ejemplos de uso
textos_prueba = [
    "La habitación estaba muy limpia y el personal era muy amable y atento",
    "El hotel era sucio y ruidoso, la experiencia fue terrible",
    "El hotel estaba bien ubicado pero la habitación era pequeña",
    "Excelente servicio, habitación espaciosa y moderna, muy cómoda y tranquila",
    "Lugar descuidado y viejo, personal lento y poco amable, muy decepcionante"
]


print("=== ANÁLISIS DE RESEÑAS HOTELERAS ===")

for texto in textos_prueba:
    resultado = analisis_sentimientos_simple(texto)
    print("\n" + "-"*50)

# Función para analizar un texto personalizado
def analizar_texto_usuario():
    while True:
        texto = input("\n\nIngrese un texto para analizar (o 'salir' para terminar): ")
        if texto.lower() == 'salir':
            break
        analisis_sentimientos_simple(texto)

print("\nPuede analizar sus propios textos:")
analizar_texto_usuario()
```
